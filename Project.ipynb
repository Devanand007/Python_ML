Math Calculator Project

```python
import math
import sympy

def add():
    a = float(input("Enter a number for a: "))
    b = float(input('Enter a number for b: '))
    print(a + b)

def subtract():
    a = float(input("Enter a number for a: "))
    b = float(input('Enter a number for b: '))
    print(a - b)

def multiply():
    a = float(input("Enter a number for a: "))
    b = float(input('Enter a number for b: '))
    print(a * b)

def divide():
    a = float(input("Enter a number for a: "))
    b = float(input('Enter a number for b: '))
    if b != 0:
        print(a / b)
    else:
        print("Division by zero is not allowed.")

def is_prime():
    number = int(input("Enter a positive integer: "))
    prime_or_comp = "prime"
    for test_number in range(2, number):
        if number % test_number == 0:
            prime_or_comp = "composite"
            break  # exit the loop as soon as we find a factor
    print(prime_or_comp)

def prime_factors():
    n = int(input('Enter a positive integer: '))
    factors = []
    divisor = 2
    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    print(factors)

def simplify_square_root():
    n = int(input('Without the radical, enter a square root to factor: '))
    upper_limit = math.floor(math.sqrt(n)) + 1
    max_factor = 1
    other_factor = 1
    square_root = 1
    for maybe_factor in range(1, upper_limit):
        if n % (maybe_factor ** 2) == 0:
            max_factor = maybe_factor ** 2
    other_factor = n / max_factor
    square_root = int(math.sqrt(max_factor))
    other_factor = int(other_factor)
    output = square_root * sympy.sqrt(other_factor)
    print(output)

def solve_variable():
    from sympy import symbols, Eq, solve

    # Define the variables
    x, y = symbols('x y')

    # Get the equations from the user
    eq1_str = input("Enter the first equation (in the form of '2*x + 3*y - 8 = 0') ignore zero: ")
    eq2_str = input("Enter the second equation (in the form of '4*x - 2*y - 2 = 0 ') ignore zero: ")

    # Parse the user input into SymPy equations
    try:
        eq1 = Eq(eval(eq1_str), 0)
        eq2 = Eq(eval(eq2_str), 0)

        # Solve the system of equations
        solution = solve((eq1, eq2), (x, y))

        # Print the solution
        print("Solution:")
        print(f"x = {solution[x]}")
        print(f"y = {solution[y]}")
    except Exception as e:
        print("Invalid input. Please enter equations in the correct format.")

# Main menu
while True:
    print("\nSelect an option:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Check if a number is prime")
    print("6. Find prime factors")
    print("7. Simplify square root")
    print("8. Solve for variables in a system of equations")
    print("9. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        add()
    elif choice == '2':
        subtract()
    elif choice == '3':
        multiply()
    elif choice == '4':
        divide()
    elif choice == '5':
        is_prime()
    elif choice == '6':
        prime_factors()
    elif choice == '7':
        simplify_square_root()
    elif choice == '8':
        solve_variable()
    elif choice == '9':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
